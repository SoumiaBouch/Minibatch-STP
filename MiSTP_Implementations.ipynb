# Minibatch Stochastic Three Points Method for Unconstrained Smooth Minimization

This is the official implementation of [Minibatch Stochastic Three Points Method for Unconstrained Smooth Minimization]().

## Requirements
We conducted the experiments using:
- python 3.8.5
- pandas 1.1.3
- numpy 1.19.2

### Datasets

For the experiments on ridge regression and regularized logistic regression problems, we used datasets from [LIBSVM](https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/) which are: abalone, a1a, australian, and splice. For the experiments on neural networks we used the [MNIST](http://yann.lecun.com/exdb/mnist/) dataset of handwritten digits.

We converted all the datsets to csv files to conduct the experiments. Those datasest are available in this repository under the names:
- abalone_scale.csv
- a1a.csv
- australian_scale.csv
- splice_scale.csv
- mnist_train.csv

## MiSTP on ridge regression and regularized logistic regression problems

To produce the results of MiSTP on ridge regression and regularized logistic regression problems follow the instructions in each section

### MiSTP with different minibatch sizes

To visualize the performance of MiSTP with different minibatch sizes on the <u> ridge regression </u> problem use the following command:

```
%run MiSTP_ridge_regression.py --dataset <input_data> --n_epochs <number_of_epochs> --n_run <number_of_runs> --minibatche_sizes <list_of_minibatche_sizes> --stepsizes <list_of_stepsizes_for_minibatches>
```

each element in the list of minibatche sizes corresponds respectively to its counterpart element in the list of stepsizes

For example, to produce the result on the abalone dataset, run the following command:

````python
%run MiSTP_ridge_regression.py --dataset abalone_scale.csv --n_epochs 30 --n_run 10 --minibatch_sizes [4177,1000,500,50,10,1] --stepsizes [1,1,1,0.1,0.01,0.001]
````

To visualize the performance of MiSTP with different minibatch sizes on the <u> regularized logistic regression </u> problem use the following command:

```
%run MiSTP_logistic_regression.py --dataset <input_data> --n_epochs <number_of_epochs> --n_run <number_of_runs> --minibatche_sizes <list_of_minibatche_sizes> --stepsizes <list_of_stepsizes_for_minibatches>
```

each element in the list of minibatche sizes corresponds respectively to its counterpart element in the list of stepsizes

For example, to produce the result on the australian dataset, run the following command:

````python
%run MiSTP_logistic_regression.py --dataset australian_scale.csv --n_epochs 70 --n_run 10 --minibatch_sizes [690,500,100,50,10] --stepsizes [0.1,0.1,0.1,0.1,0.01]
````

### MiSTP vs. Other Zero Order methods

To produce the results of comparison of the perfomance of MiSTP, RSGF, ZO_SVRG, and ZO_CD on the <u> ridge regression </u> problem run the following command:

``` ridge_others
%run MiSTP_vs_others_ridge.py --n_epochs <number_of_epochs> --n_run <number_of_runs> --batch_size <batch_size> --alpha_MiSTP <stepsize_for_MiSTP> --alpha_RSGF <stepsize_for_RSGF> --alpha_ZO_SVRG <stepsize_for_ZO_SVRG> --alpha_ZO_CD <stepsize_for_ZO_CD>
```

For example, to run this command using a minibatch size of 50 we can use:

````python
%run MiSTP_vs_others_ridge.py --n_epochs 100 --n_run 3 --batch_size 50  --alpha_MiSTP 0.1 --alpha_RSGF 0.1 --alpha_ZO_SVRG 0.001 --alpha_ZO_CD 0.1
````

To produce the results of comparison of the perfomance of MiSTP, RSGF, ZO_SVRG, and ZO_CD on the <u> logistic regression </u> problem run the following command:

``` ridge_others
%run MiSTP_vs_others_logistic.py --n_epochs <number_of_epochs> --n_run <number_of_runs> --batch_size <batch_size> --alpha_MiSTP <stepsize_for_MiSTP> --alpha_RSGF <stepsize_for_RSGF> --alpha_ZO_SVRG <stepsize_for_ZO_SVRG> --alpha_ZO_CD <stepsize_for_ZO_CD>
```

For example, to run this command using a minibatch size of 100 we can use:

````python
%run MiSTP_vs_others_logistic.py --n_epochs 300 --n_run 1 --batch_size 100  --alpha_MiSTP 0.1 --alpha_RSGF 1 --alpha_ZO_SVRG 0.01 --alpha_ZO_CD 1
````

## MiSTP in neural networks

To produce the results of MiSTP on neural networks run the following command:

```
%run MiSTP_neural_networks.py --dataset <input_data> --n_epochs <number_of_epochs> --n_run <number_of_runs> --minibatche_sizes <list_of_minibatche_sizes> --stepsizes <list_of_stepsizes_for_minibatches>
```

each element in the list of minibatche sizes corresponds respectively to its counterpart element in the list of stepsizes

For example, the following command produce the results of one run of MiSTP, using the full batch, to classify MNIST data

````python
%run MiSTP_neural_networks.py --dataset mnist_train.csv --n_epochs 4000 --n_run 1 --minibatch_sizes [60000] --stepsizes [0.1]
````

### Generation of a black box adversarial attack on a deep neural network classifier

For this experiments we used the codes from https://github.com/IBM/ZOSVRG-BlackBox-Adv , we added just the code for MiSTP to the optimization_methods folder in the same manner done for ZO-SGD and ZO-SVRG. The file Minibatch_STP.py for this is available in this repository.

### MiSTP vs. SGD

To produce the results of MiSTP & SGD on the <u> ridge regression </u> problem run the following command:

``` ridge
%run MiSTP_vs_SGD_ridge.py --dataset <input_data> --n_iter <number_of_iterations> --n_run <number_of_runs> --batch_size <batch_size> --alpha_MiSTP <stepsize_for_MiSTP> --alpha_SGD <stepsize_for_SGD>
```

For example, to run this command using the abalone dataset and for a minibatch size of 100 we use:

````python
%run MiSTP_vs_SGD_ridge.py --dataset abalone_scale.csv --n_iter 30 --n_run 10 --batch_size 100 --alpha_MiSTP 1 --alpha_SGD 1 
````

To produce the results of MiSTP & SGD on the <u> regularized logistic regression </u> problem run the following command:

``` ridge
%run MiSTP_vs_SGD_logistic.py --dataset <input_data> --n_iter <number_of_iterations> --n_run <number_of_runs> --batch_size <batch_size> --alpha_MiSTP <stepsize_for_MiSTP> --alpha_SGD <stepsize_for_SGD>
```

For example, to run this command using the a1a dataset and for a minibatch size of 100 we use:

````python
%run MiSTP_vs_SGD_logistic.py --dataset a1a.csv --n_iter 2000 --n_run 10 --batch_size 100 --alpha_MiSTP 0.01 --alpha_SGD 1 
````
